```python
df_train_users = df_train.groupby('user_id').agg('max')
df_train_users
```

```python
# macierz do grupowania ludzi w grupy demograficzne
df_train_users_dummies = pd.get_dummies(df_train_users[['bust size', 'weight', 'body type', 'height', 'age']],
                                        columns=['bust size', 'weight', 'body type', 'height', 'age'])
df_train_users_dummies
```

```python
#  macierz potrzebna, aby znaleźć najpopularniejsze przedmioty w danej grupie demograficznej
pivot_user_ratings = df_train_users.pivot_table(values="rating", 
                                               columns="item_id", 
                                               index="user_id", 
                                               aggfunc="count", 
                                               fill_value=0)

pivot_user_ratings
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import KMeans

user_demographic_cosine = cosine_similarity(df_train_users_dummies, df_train_users_dummies)

def k_select(k):
    '''
    function to find optimal number of clusters for cosine similiarity matrix
    k: the number of cluster in the k-means solution
    output is dictionary with k and sum of square distance for cluster solution
    '''
    
    rating_kmeans = KMeans(n_clusters = k, max_iter = 400, 
                           algorithm = "auto", n_init = 30, random_state = 1017).fit(user_demographic_cosine)
    ssq_dist = {"cluster_n": k,
                "ssq_dist": rating_kmeans.inertia_}
    return ssq_dist
```

```python
from tqdm import tqdm
k_options = list(range(2, 15 + 1))
kmean_ssq_dist = []
for k in tqdm(k_options):
    kmean_ssq_dist.append(k_select(k))
# kmean_ssq_dist = pd.DataFrame([k_select(k) for k in k_options])
pd.DataFrame(kmean_ssq_dist["ssq_dist"].diff()[1:])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 7))

plt.plot(kmean_ssq_dist["cluster_n"], kmean_ssq_dist["ssq_dist"], color="orange", linewidth = 5)
plt.xticks(k_options, k_options)
plt.xlabel("K selection")
plt.ylabel("Inertia (within-cluster sum-of-squares)")
plt.title("Selecting k = 10 looks like a reasonable choice given variance plateau")
plt.show()
```

```python
rating_kmeans = KMeans(n_clusters = 10, max_iter = 400, 
                       algorithm = "auto", n_init = 30, 
                       random_state = 1017).fit(user_demographic_cosine)
user_clusters = pd.DataFrame(rating_kmeans.labels_, columns=["cluster_n"])
pd.DataFrame(user_clusters["cluster_n"].value_counts())
```

```python
pivot_user_ratings["cluster"] = list(rating_kmeans.labels_)
pivot_user_ratings[["cluster"]].head()
```

```python
cluster_items_melt = pivot_user_ratings.melt(id_vars="cluster", value_name="rating")
cluster_items_melt = pd.DataFrame(cluster_items_melt.groupby(["cluster", "item_id"])["rating"].mean())
cluster_items_melt.reset_index(inplace=True)
cluster_items_melt.sort_values(["cluster", "rating"], ascending=False, inplace=True)

cluster_items_melt
```

```python
def get_cluster_items(cluster_id):
    items_ratings = cluster_items_melt[
    cluster_items_melt['cluster'] == cluster_id][['item_id','rating']].sort_values('rating',
                                                                          ascending = False).head(120).to_records(index=False)
    return {str(it): rat for it, rat in list(items_ratings)}
```

```python
def get_user_items(user_id):
    cluster_id = pivot_user_ratings['cluster'][user_id]
    cluster_items = get_cluster_items(cluster_id)
    return cluster_items
list(get_user_items(332).items())[:5]
```
